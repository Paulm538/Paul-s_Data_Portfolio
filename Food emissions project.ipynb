# Cleaning and Merging of Datasets CO2 Emission Production by Country, Kcal of Food Consumed and Population

Next, in order to calculate how much CO2 could be saved by not consuming certain products one must add the following datasets. Population size by  country, amount of kcal consumed by a person per day per country and the CO2 emissions produced by each country. In order to imitate a scientific experiment. The above date sets will act as the control variables and thus, all data will be from the year 2014 when metering the datasets together for calculations. This will done to keep our results from the model as accurate and useful as possible given the limitations in the datasets as a whole.    

Step One : 

Will be to change the column names for year and country so that all these names are same in order to merge the datasets together on these columns. 

```python
# Notes: 
#       cdf_2 = uncleaned date frame
#        s1_df_2 = stage one of procssing the data frame
#       s2_df_2 =  stage two of procssing  the data frame
#       fv_df_2 = cleaned data frame ready to be merged


cdf_2 = df_2.rename(columns = {'country_or_area' : 'country'})
s1_df_2 = cdf_2.loc[:, ['country', 'year', 'value', 'category']]
s2_df_2 = s1_df_2[s1_df_2['year'] == 2014]
fv_df_2 = s2_df_2[s2_df_2['category'] == 'carbon_dioxide_co2_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent']
```

```python
# Notes: 
        # cdf_3 = uncleaned date frame
        # s1_df_3 = stage one of procssing the data frame
        # fv_df_3 = cleaned data frame ready to be merged

cdf_3 = df_3.rename(columns = {'Entity' : 'country', 'Year' : 'year', 'Food supply (kcal/capita/day) (FAO, 2020)' : 'kcal/pre/day'})

s1df_3 = cdf_3.loc[:, ['country', 'year', 'kcal/pre/day' ]]

fv_df_3 = s1df_3[s1df_3['year'] == 2014]
```

```python
# Notes: 
        # cdf_4 = uncleaned date frame
        # s1_df_4 = stage one of procssing the data frame
        # fv_df_4 = cleaned data frame ready to be merged


cdf_4 = df_4.rename(columns = {'Entity' : 'country', 'Year' : 'year', 'Population by country and region, historic and projections (Gapminder, HYDE & UN)' : 'population'})

s1df_4 = cdf_4.loc[:, ['country', 'year', 'population' ]]

fv_df_4 = s1df_4[s1df_4['year'] == 2014]
```

Step Two: 

Will be to merge the three datasets together and then delete all duplicate year columns as they are no longer needed. Also changing the value column name to co2_adm (co2 emissions) for easier reading during calculation stage.

Notes:
mdf = merged date frame
s1_ mdf = Stage one of procssing the merged data frame
fv_ mdf = final version of the merged data frame ready for calculations

```python
mdf = pd.merge(pd.merge(fv_df_2,fv_df_3, on= 'country'),fv_df_4, on= 'country')

s1mdf = mdf.drop(['year_y', 'category' , 'year'], axis=1)
fv_mdf = s1mdf.rename(columns = {'year_x' : 'year', 'value' : 'co2_adm' })
```

# Calculations

Here the calculates will be performed to try and model how much CO2 emission a country will save in the year 2014 if its entire population would stop consuming certain products from one or more food packages.

Note: 

(0.00013) Unit conversion of Kcal to KG


'kg_of_food_per_day' =  amount of food one person consumes each day from a specific country in the year 2014

'co2_per_day_per_person(kg CO2e)' = amount of co2_emissions create from the Food package or packages by one person from a specific country in the year 2014

'Co2_produced_country_population(kT CO2e)' = total co2 as (kT CO2e) produced by a Food package or packages per day by whole population of a specific country in the year 2014

'total_co2_reduction_as_precentage(kT CO2e)'= total co2e reduction of a country by whole population per day of a specific country in the year 2014 Than converted to a percent (%)


# Reduction of CO2 where a Population stops eating meat (per day)

Note: 

t_c02_mdm = meat food package
t_c02_mdveg2 = unsustainable food package

non_meat = amount of Co2e reduced from not consuming any meat

un_meat = amount of Co2e reduced from not consuming any meat and stops consuming products from the unstable food package

```python
fv_mdf['kg_of_food_per_day'] = fv_mdf['kcal/pre/day'] * 0.00013 

fv_mdf['co2_per_day_per_person(kg CO2e)'] = fv_mdf['kg_of_food_per_day'] * t_c02_mdm['Total Global Average GHG Emissions per kg']

fv_mdf['Co2_produced_country_population(kT CO2e)'] = (fv_mdf['co2_per_day_per_person(kg CO2e)'] * fv_mdf['population'])/1000000


fv_mdf['total_co2_reduction_as_precentage(kT CO2e)'] = (fv_mdf['Co2_produced_country_population(kT CO2e)'] / fv_mdf['co2_adm'])*100

#HTML(fv_mdf.to_html(classes = 'table table-striped'))


non_meat = fv_mdf[['country','total_co2_reduction_as_precentage(kT CO2e)']]
non_meat = non_meat.rename(columns = {'total_co2_reduction_as_precentage(kT CO2e)' : 'non_Meat_co2_reduction_as_%(kT CO2e)'})

```

```python
fv_mdf['kg_of_food_per_day'] = fv_mdf['kcal/pre/day'] * 0.00013 

fv_mdf['co2_per_day_per_person(kg CO2e)'] = fv_mdf['kg_of_food_per_day'] * (t_c02_mdm['Total Global Average GHG Emissions per kg'] + t_c02_mdveg2[ 'Total Global Average GHG Emissions per kg'])

fv_mdf['Co2_produced_country_population(kT CO2e)'] = (fv_mdf['co2_per_day_per_person(kg CO2e)'] * fv_mdf['population'])/1000000


fv_mdf['total_co2_reduction_as_precentage(kT CO2e)'] = (fv_mdf['Co2_produced_country_population(kT CO2e)'] / fv_mdf['co2_adm'])*100

#HTML(fv_mdf.to_html(classes = 'table table-striped'))

un_meat = fv_mdf[['country','total_co2_reduction_as_precentage(kT CO2e)']]
un_meat = un_meat.rename(columns = {'total_co2_reduction_as_precentage(kT CO2e)' : 'UnSt_Meat_co2_reduction_as_%(kT CO2e)'})
```

# Reduction of CO2 where a Population have a Vegan (per day)

```python
fv_mdf['kg_of_food_per_day'] = fv_mdf['kcal/pre/day'] * 0.00013 

fv_mdf['co2_per_day_per_person(kg CO2e)'] = fv_mdf['kg_of_food_per_day'] * (t_c02_mdveg2[ 'Total Global Average GHG Emissions per kg'] + t_c02_mdveg1['Total Global Average GHG Emissions per kg'] + t_c02_mdm['Total Global Average GHG Emissions per kg'])


fv_mdf['Co2_produced_country_population(kT CO2e)'] = (fv_mdf['co2_per_day_per_person(kg CO2e)'] * fv_mdf['population'])/1000000

fv_mdf['total_co2_reduction_as_precentage(kT CO2e)'] = (fv_mdf['Co2_produced_country_population(kT CO2e)'] / fv_mdf['co2_adm'])*100 

#HTML(fv_mdf.to_html(classes = 'table table-striped'))

un_vegan = fv_mdf[['country','total_co2_reduction_as_precentage(kT CO2e)']]
un_vegan = un_vegan.rename(columns = {'total_co2_reduction_as_precentage(kT CO2e)' : 'UnSt_Vegan_co2_reduction_as_%(kT CO2e)'})
```
